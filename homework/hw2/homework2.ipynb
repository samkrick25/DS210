```python
#read data into DataFrame
import pandas as pd

filename = 'student-mat.csv'
data = pd.read_csv(filename, sep=';')
data.dtypes
```

```python
#create a DataFrame containing the features I want to use, then checking datatypes again to make sure everything lines up

data1 = data[['Medu','Fedu','studytime','schoolsup','paid','absences','Mjob','Fjob','higher',
              'goout','Dalc','famrel','failures','health','freetime','famrel'
]]
#data1 = data[['absences','Medu','Fedu','studytime','famrel']]
#create my target set
target = data['G3']
data1.dtypes
```

```python
#one hot encode categorical columns to get boolean values, create an array that has a boolean value for each unique category
from sklearn.preprocessing import OneHotEncoder #GPT gen code
catdata = data1[['schoolsup', 'paid', 'Mjob', 'Fjob', 
                 'higher'
]]
encoder = OneHotEncoder(sparse_output=False)
encodings = encoder.fit_transform(catdata) #end of GPT gen code
data1 = data1.drop(['schoolsup', 'paid', 'Mjob', 'Fjob', 
                    'higher'], axis=1
) #drop non one hot encoded columns
usedata = np.hstack((data1, encodings)) #horizontally concatenate one hot encodings with the rest of my dataset
usedata
```

```python
#whiten data so that STD's make sense with each other, all data has similar ranges
from scipy.cluster.vq import whiten
usedata = whiten(usedata)
usedata
```

```python
#split into training and test sets
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(usedata, target)
```

```python
#for loop to iterate over a range of clusters, will use the best results for my testing. This will output a nested dictionary
#in which the keys are first the number of clusters used, then each cluster, then the assignment of students to each cluster using vq
from scipy.cluster.vq import kmeans, vq
x_train = x_train.astype(np.float32)
allassign = {}
centroids = {}   #to be used to calculate distance to test set points
for c in range(2,21):
    clustered,_ = kmeans(x_train, c)
    centroids[c] = clustered
    assigned,_ = vq(x_train, clustered)
    allassign[c] = assigned
```

```python
'''
create a nested dictionary that holds first each separate value of k that I used earlier, then each unique cluster, with the values for the 
final keys being the most popular grade in each cluster. This code was inspired by a conversation I had with chatGPT which is linked below.
'''
from scipy.stats import mode
clustgradebyk = {}  #create the end nested dictionary
for k, assign in allassign.items():  #iterate over my grade assignments
    gradeassign = {}  #dictionary where keys will be each cluster, and values will be the most popular grade in that cluster
    for cluster in np.unique(assign):
        ind = cluster==assign  #get indices of each point assigned to each unique cluster, given in boolean values
        clustgrade = y_train[ind] #where ind is true, means that that student in the training set was clustered into the cluster I am iterating over, store all of these grades
        grademode = mode(clustgrade)[0] #find most common grade per cluster, then only access the mode and not the count
        gradeassign[cluster] = grademode #place the mode as the value to the key of each cluster I iterate over, so that each key in gradeassign is a different cluster and the value is the mode grade in that cluster
    clustgradebyk[k] = gradeassign #place this dictionary as the value connected to the key in the first dictionary, which is each value of k I iterate over
for k, clust in clustgradebyk.items():  #print everything out so I can see the mode grade by cluster by k
    print(f'for k = {k}:')
    for c, grade in clust.items():
        print(f'Cluster {c+1} Mode: {grade}')
```

```python
'''
some functions used here are described in the notes. This code block will output a graph of accuracy metrics, after assigning my test set to clusters. 
This code was written in similar fashion to the code evaluating the training set, which was inspired by chatGPT from the same conversation 
shown in the notes below
'''
from sklearn.metrics import mean_squared_error
error = {}
for k, cent in centroids.items():
    distances = cdist(x_test, cent)
    nearestcent = np.argmin(distances, axis=1)
    predicted_grades = np.zeros(len(x_test))  #initiate a matrix of zeroes to store predicted grades, done with 0s so that indices can be used to store values later on
    for cluster, grade in clustgradebyk[k].items():
        clustind = nearestcent == cluster
        predicted_grades[clustind] = grade
    mse = mean_squared_error(y_test, predicted_grades)
    error[k] = mse
    print(f'MSE for k = {k}: {mse}')
import matplotlib.pyplot as plt
kvals = list(error.keys())
msevals = list(error.values())
plt.plot(kvals, msevals, marker='o')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Mean Squared Error (MSE)')
plt.title('Accuracy by Number of Clusters')
plt.grid(True)
plt.show()
```
