# Question 3

```python
import numpy as np
import matplotlib.pyplot as plt
def f(x):
    return 0.5 + 0.5 * np.sin(2*np.pi*x)  #sinusoid function starting at 0.5 and outputting values in the range [0,1], while going through a full period in that distance
# Generate data points
X = np.linspace(0, 1, 100)
y = f(X)
```

```python
from sklearn.tree import DecisionTreeRegressor
#train a decision tree using MSE
mse_tree = DecisionTreeRegressor(criterion='squared_error', max_leaf_nodes=6)
mse_tree.fit(X.reshape(-1, 1), y)
#train a decision tree using MAE
mae_tree = DecisionTreeRegressor(criterion='absolute_error', max_leaf_nodes=6)
mae_tree.fit(X.reshape(-1, 1), y)
```

```python
# Plot the function f(x), MSE prediction, and MAE prediction
plt.figure(figsize=(10, 6))
plt.plot(X, y, label='f(x)', color='blue')
plt.plot(X, mse_tree.predict(X.reshape(-1, 1)), label='MSE Prediction', color='green')
plt.plot(X, mae_tree.predict(X.reshape(-1, 1)), label='MAE Prediction', color='orange')
plt.xlabel('x')
plt.ylabel('y')
plt.title('F(x) and predictions with MAE and MSE')
plt.grid(True)
plt.legend()
plt.show()
```

MSE and MAE result in very similar loss functions, as seen on the graph, however there are some slight differences. Using MSE, the goal of the loss function is to minimize the squared differences between predicted values and actual values. Because of this, the MSE function will be more sensitive to larger outliers, and may capture some noise instead of fitting to the curve smoother. This is seen in the peak and valley of this function, where the MAE prediction is closer to the peak or vallue. MAE only takes the absolute difference between actual and predicted values, without squaring. This means that it is not sensitive to outliers, but will also capture less noise and fit the function more smoothly. This is also seen in the middle of the graph, where the MAE function changes quicker and follows the function better than the MSE function.

# Question 4

```python
#generate noisy data
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(42)
def f_x(x):
    return np.sin(2 * np.pi * x)
x = np.linspace(0, 1, 100)
y_noisy = f_x(x) + np.random.normal(0,0.2,len(x))
plt.scatter(x, y_noisy)
```

```python
degrees = [1, 3, 10, 20]
plt.figure(figsize=(15, 10))
for i, degree in enumerate(degrees):
    coeffs = np.polyfit(x, y_noisy, degree)
    y_pred = np.polyval(coeffs, x)    #polyval does what eval_poly does from lecture 09, evaluating the polynomial at specific values for x
    error = np.mean((y_pred - y_noisy) ** 2)  #calculating MSE here, to be used in the legend label
    plt.subplot(2, 2, i+1)
    plt.scatter(x, y_noisy, label='Noisy Data')
    plt.plot(x, y_pred, 'r-', label=f'Degree {degree} (Error: {error:.3f})')
    plt.plot(x, f_x(x), 'g-', label='True Function')
    plt.title(f'Degree {degree}')
    plt.legend()
plt.show()
```

Here, you can see the function underfit at degree 1, not being able to capture the curve of the data when trying to use a linear function to represent it. At degree 3, the polyfit object is very close to the true function, obviously being slightly affected by noise. You can begin to see evidence of overfitting when using a polynomial of degree 10, but it still stays somewhat close to the function. At degree 20, overfitting is very clear, as you can see the polyfit object move and curve with respect to the noise in the data. You can also see the MSE displayed in the legend for each degree, and it increases as the degree increases, also showing overfitting.
